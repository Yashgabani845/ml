```python
import pandas as pd

# Sample DataFrame with missing values
data = {
    'Feature1': [1, 2, 3, None, 5],
    'Feature2': [10, None, 30, 40, 50],
    'Feature3': [None, 2, 3, 4, 5]
}
df = pd.DataFrame(data)

# Remove rows with any missing value
df_cleaned_rows = df.dropna()

# Remove columns with any missing value
df_cleaned_columns = df.dropna(axis=1)

print("Rows cleaned:")
print(df_cleaned_rows)

print("\nColumns cleaned:")
print(df_cleaned_columns)
```

```python
from sklearn.impute import SimpleImputer
import pandas as pd

# Sample DataFrame with missing values
data = {
    'Feature1': [1, 2, 3, None, 5],
    'Feature2': [10, None, 30, 40, 50],
    'Feature3': [None, 2, 3, 4, 5]
}
df = pd.DataFrame(data)

# Initialize imputer for numeric data (using mean)
imputer = SimpleImputer(strategy='mean')

# Apply imputer to fill missing values
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print("Imputed DataFrame:")
print(df_imputed)
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Sample data with varying scales
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [30000, 50000, 70000, 90000, 110000]
}
df = pd.DataFrame(data)

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Scale the data
scaled_data = scaler.fit_transform(df)

# Convert the scaled data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("Original DataFrame:")
print(df)

print("\nScaled DataFrame (Min-Max Normalization):")
print(scaled_df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Sample data with varying scales
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [30000, 50000, 70000, 90000, 110000]
}
df = pd.DataFrame(data)

# Perform Mean Normalization
df_normalized = (df - df.mean()) / (df.max() - df.min())

print("Original DataFrame:")
print(df)

print("\nMean Normalized DataFrame:")
print(df_normalized)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Sample data with varying scales
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [30000, 50000, 70000, 90000, 110000]
}
df = pd.DataFrame(data)

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the data
scaled_data = scaler.fit_transform(df)

# Convert the standardized data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("Original DataFrame:")
print(df)

print("\nStandardized DataFrame (Z-Score Normalization):")
print(scaled_df)
```

```python
from sklearn.preprocessing import Binarizer
import pandas as pd

# Sample data
data = {
    'Salary': [30000, 50000, 70000, 90000, 110000]
}
df = pd.DataFrame(data)

# Initialize the Binarizer with a threshold of 50000
binarizer = Binarizer(threshold=50000)

# Binarize the data
binary_data = binarizer.fit_transform(df)

# Convert the binary data back to a DataFrame
binary_df = pd.DataFrame(binary_data, columns=df.columns)

print("Original DataFrame:")
print(df)

print("\nBinarized DataFrame:")
print(binary_df)
```

```python
# prompt: apply preporcessing on that datasert

import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Binarizer, LabelEncoder

try:
    df = pd.read_csv('Cardata.csv')
except FileNotFoundError:
    print("Error: 'Cardata.csv' not found. Please ensure the file exists in the current directory or provide the correct path.")
    exit()

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(exclude=['number']).columns

# Handle missing values in numerical features
imputer_numerical = SimpleImputer(strategy='mean')  # or 'median', 'most_frequent'
df[numerical_cols] = imputer_numerical.fit_transform(df[numerical_cols])

# Handle missing values in categorical features
imputer_categorical = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = imputer_categorical.fit_transform(df[categorical_cols])


# Encode categorical features using Label Encoding
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Store the encoder for potential inverse transformation later


# Feature Scaling (choose one or combine as needed)

# 1. Min-Max Scaling
scaler_minmax = MinMaxScaler()
df[numerical_cols] = scaler_minmax.fit_transform(df[numerical_cols])

#scaler_standard = StandardScaler()
#df[numerical_cols] = scaler_standard.fit_transform(df[numerical_cols])

print(df.head())
```

```python
# prompt: generate correlation graph for this

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df = pd.read_csv('Cardata.csv')
except FileNotFoundError:
    print("Error: 'Cardata.csv' not found. Please ensure the file exists in the current directory or provide the correct path.")
    exit()

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(exclude=['number']).columns

# Handle missing values in numerical features
imputer_numerical = SimpleImputer(strategy='mean')  # or 'median', 'most_frequent'
df[numerical_cols] = imputer_numerical.fit_transform(df[numerical_cols])

# Handle missing values in categorical features
imputer_categorical = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = imputer_categorical.fit_transform(df[categorical_cols])

# Encode categorical features using Label Encoding
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Store the encoder for potential inverse transformation later

# Feature Scaling (choose one or combine as needed)
scaler_minmax = MinMaxScaler()
df[numerical_cols] = scaler_minmax.fit_transform(df[numerical_cols])


# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plotting the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Car Data')
plt.show()
```
